```python
# pre_process.py
import re
import json
from pathlib import Path
from typing import Dict, Optional, List


def read_txt(path: str) -> str:
    text = Path(path).read_text(encoding="utf-8")
    return text.lstrip("\ufeff")  # 去掉 BOM


def clean_text(text: str) -> str:
    # 将换行转为空格
    text = text.replace("\r\n", "\n").replace("\r", "\n")
    lines = [line.strip() for line in text.split("\n")]
    lines = [line for line in lines if line]
    return " ".join(lines)


def split_sentences(text: str) -> List[str]:
    # 按常见中文句末分句，保留标点
    pattern = r"(?<=[。！？!?])\s+"
    raw_sents = re.split(pattern, text)
    return [s.strip() for s in raw_sents if s.strip()]


def preprocess_file(path: str, save_json: Optional[str] = None) -> Dict[str, str]:
    """
    返回结构：
    {
        "000001": "句子1",
        "000002": "句子2",
        ...
    }
    """
    text = read_txt(path)
    text = clean_text(text)
    sents = split_sentences(text)

    # 生成 {id: sentence}
    sent_dict = {
        f"{i:05d}": s for i, s in enumerate(sents, start=1)
    }

    if save_json:
        json.dump(sent_dict, open(save_json, "w", encoding="utf-8"),
                  ensure_ascii=False, indent=2)

    return sent_dict


if __name__ == "__main__":
    import argparse

    parser = argparse.ArgumentParser()
    parser.add_argument("input")
    parser.add_argument("--output", "-o", default=None)
    args = parser.parse_args()

    sent_dict = preprocess_file(args.input, save_json=args.output)
    print(f"共分句：{len(sent_dict)}")
    for k, v in list(sent_dict.items())[:5]:
        print(k, v)
```
